```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mart = pd.read_excel("supermarket_sales.xlsx") # DataFrame object
mart.head()
```

```python
mart.tail()
```

```python
plt.figure(figsize=(15,5))

sns.countplot(x="Product line",data=mart)
```

```python
plt.figure(figsize=(15,5))

sns.countplot(x="Gender",data=mart)
```

```python
plt.figure(figsize=(15,5))

sns.countplot(x="Product line",hue="Gender",data=mart)
```

```python
plt.figure(figsize=(15,5))

sns.countplot(x="Product line",hue="Gender",data=mart,palette='PuBuGn_r')
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",hue="Gender",data=mart,palette='YlGnBu')
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",hue="Gender",data=mart,palette='YlGnBu', stat='count', width=0.8, saturation=0.75)
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x="Product line",hue="Gender",data=mart,palette='YlGnBu', stat='percent', width=0.8, saturation=0.75)
```
